Cafe

```python
import pandas as pd

# Read the dataset
df = pd.read_csv("D:\\Internship Progress\\datasets\\datasets\\food and beverages\\cafe.csv")

# Print the column names to identify the correct column name
print(df.columns)

# Replace 'ColumnName' with the actual column name from the printed list
df['full_text'] = df['full_text'].str.lower()  # Convert to lowercase
df['full_text'] = df['full_text'].str.replace(r'[^a-zA-Z\s]', '')  # Remove non-alphabetic characters
df['full_text'] = df['full_text'].str.replace(r'\s+', ' ')  # Remove extra spaces
df['full_text'] = df['full_text'].str.replace(r'http\S+|www\S+', '', case=False)  # Remove words starting with http or www
# Print the cleaned tweets
pd.set_option('display.max_colwidth', None)
print(df)
df.to_csv("D:\\Internship Progress\\datasets\\datasets\\food and beverages\\cafe.csv")
```
